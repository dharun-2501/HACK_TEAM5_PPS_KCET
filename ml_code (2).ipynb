```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
sal=pd.read_csv("/content/dataset.csv")
```

```python
sal.head(10)
```

```python
sal.isnull().sum()
```

```python
le=LabelEncoder()
le2=LabelEncoder()
```

```python
sal.columns = sal.columns.str.strip()
```

```python

sal['shop_freq']=le.fit_transform(sal['Monthly Shopping frequency'])
sal['data_reg']=le2.fit_transform(sal['How much do you usually spend on mobile data recharge per month?'])
```

```python
sal['Amount'] = np.random.randint(10000, 50000, size=len(sal))
```

```python
sal.head(5)
```

```python
indep=sal[['Number of outings per month','shop_freq','data_reg','How often do you order food online (e.g., Zomato, Swiggy) in a month?','How much do you pay for rent or hostel accommodation per month?']]
dep=sal['Amount']
```

```python
numeric_cols_to_convert = ['Number of outings per month', 'How often do you order food online (e.g., Zomato, Swiggy) in a month?','How much do you pay for rent or hostel accommodation per month?']
```

```python
# Convert the specified columns to numeric, handling potential commas
for col in numeric_cols_to_convert:
    # Replace comma with period and convert to numeric, coercing errors to NaN
    indep[col] = indep[col].astype(str).str.replace(',', '.', regex=False)
    indep[col] = pd.to_numeric(indep[col], errors='coerce')

# Handle any NaN values that resulted from coercion, e.g., by filling with the mean
indep = indep.fillna(indep.mean())

# Now fit the model with the cleaned data
lr=LinearRegression()
lr.fit(indep,dep)
```

```python
print("the total amount is ",lr.predict([[3,1,2,3,1000]]))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)


print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```
